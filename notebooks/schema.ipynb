## categorical

```python
def view(data):
    data_info = pd.DataFrame({"missing":data.isna().sum(),
    "missing(%)": np.round(data.isna().sum()/data.shape[0]*100, 2),
    "unique":data.nunique(),
    "avail": data.notna().sum()
    })
    return data_info

view(categorical_df)["unique"]
```

```python
categ_info = view(categorical_df)
categ_drop = categ_info[categ_info["missing(%)"] >= 45].index
categorical_df = categorical_df.drop(list(categ_drop), axis=1)
```

```python
cat_schema = pd.Series(data=[categorical_df[i].unique() for i in categorical_df.columns],
    index=[i for i in categorical_df.columns])
cat_schema.to_json()
```
